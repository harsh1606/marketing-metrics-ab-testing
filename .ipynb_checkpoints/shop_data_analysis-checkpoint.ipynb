## Frequentist A/B Test  CTR & ROAS (Variant A vs B)

```python
# A/B aggregate: Variant A vs B

ab_agg = shop_df.groupby("variant", observed=True).agg(
    impressions=("Impressions", "sum"),
    clicks=("Clicks", "sum"),
    conversions=("Conversions", "sum"),
    cost=("Cost", "sum"),
    revenue=("Revenue", "sum")
).reset_index()

ab_agg["ctr"]  = ab_agg["clicks"] / ab_agg["impressions"]
ab_agg["cvr"]  = ab_agg["conversions"] / ab_agg["clicks"].replace(0, np.nan) 
ab_agg["roas"] = ab_agg["revenue"] / ab_agg["cost"].replace(0, np.nan)

ab_agg
```

### Frequentist A/B Summary: Variant A vs B

Here I split campaigns into two variants:
- **Variant A** = coupon/discountoriented ad groups  
- **Variant B** = all other campaigns  

From the aggregated metrics:

- Variant A gets **more impressions and clicks** and has a **higher CTR (~36.8% vs ~31.1%)**, so coupon ads are better at attracting traffic.
- Variant B has a **higher conversion rate per click (~13.6% vs ~8.6%)**, meaning its clicks are higher quality.
- In total revenue, A is higher (because of more volume), but **ROAS is slightly worse** for A (0.87) than B (0.90), and both are below 1, so neither segment is profitable on ad spend.


This sets up the A/B testing questions:
1. Is the CTR difference between A and B statistically significant?
2. Is the ROAS difference between A and B statistically significant?
3. Are coupon campaigns just buying cheap clicks, or are they actually good for profitability?

## Two-proportion z-test on CTR (clicks / impressions)

```python
from statsmodels.stats.proportion import proportions_ztest

# Make sure we align rows for A and B
a_row = ab_agg[ab_agg["variant"] == "A"].iloc[0]
b_row = ab_agg[ab_agg["variant"] == "B"].iloc[0]

counts = np.array([a_row["clicks"], b_row["clicks"]])          # successes = clicks
nobs   = np.array([a_row["impressions"], b_row["impressions"]])  # trials   = impressions

z_stat, p_val = proportions_ztest(count=counts, nobs=nobs, alternative="two-sided")

print(f"CTR A (coupon/discount) = {a_row['ctr']:.4%}")
print(f"CTR B (non-coupon)      = {b_row['ctr']:.4%}")
print(f"z-stat = {z_stat:.3f}, p-value = {p_val:.4f}")
```

### Two-Proportion Z-Test on CTR (Variant A vs B)

Here I compare click-through rate (CTR) between:

- **Variant A** = coupon/discount campaigns  
- **Variant B** = non-coupon campaigns  

Hypotheses:

- **H:** p_A = p_B  no difference in CTR  
- **H:** p_A  p_B  CTR differs between groups  

Using a two-proportion z-test:

- CTR(A)  36.8%  
- CTR(B)  31.1%  
- z  94.98, p-value  0.0000  

Because the p-value is effectively zero, I **reject H** and conclude that coupon/discount campaigns have a **statistically significantly higher CTR** than non-coupon campaigns.

However, this is only about **click probability**. In later steps I compare **conversion rate and ROAS** to see whether these extra clicks are actually profitable.


## 95% CI for difference in CTR (B - A)

```python
p_A = a_row["ctr"]
p_B = b_row["ctr"]
n_A = a_row["impressions"]
n_B = b_row["impressions"]

diff = p_B - p_A

se_diff = np.sqrt(
    p_A * (1 - p_A) / n_A +
    p_B * (1 - p_B) / n_B
)

z_crit = 1.96  # for ~95% CI
ci_low  = diff - z_crit * se_diff
ci_high = diff + z_crit * se_diff

print(f"Difference (B - A) in CTR = {diff:.4%}")
print(f"95% CI for diff = [{ci_low:.4%}, {ci_high:.4%}]")
```

### 95% Confidence Interval for CTR Difference (B  A)

I computed a 95% confidence interval for the difference in CTR between:

- **Variant A** = coupon/discount campaigns (CTR = p_A)  
- **Variant B** = non-coupon campaigns (CTR = p_B)  

The parameter of interest is:

\[
\Delta = p_B - p_A
\]

Using the standard error for the difference of two independent proportions and a 1.96 z-critical value for 95% CI, I obtained:

- Difference (B  A) in CTR  **5.66 percentage points**
- 95% CI = **[5.78%, 5.55%]**

The entire interval is below 0, which confirms the z-test result:  
**non-coupon campaigns have a significantly lower CTR than coupon/discount campaigns**, by about **5.6 percentage points** on average.

This quantifies the effect size: coupon/discount ads are reliably better at getting clicks, although later analyses (CVR, ROAS, P&L) check whether those extra clicks are actually profitable.


## Two-proportion z-test on Conversion Rate (CVR = conversions / clicks)

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# Reuse rows for A and B from ab_agg
a_row = ab_agg[ab_agg["variant"] == "A"].iloc[0]  # coupon/discount
b_row = ab_agg[ab_agg["variant"] == "B"].iloc[0]  # non-coupon

# Successes = conversions, Trials = clicks
conv_counts = np.array([a_row["conversions"], b_row["conversions"]])
clicks_nobs = np.array([a_row["clicks"],       b_row["clicks"]])

z_cvr, p_cvr = proportions_ztest(
    count=conv_counts,
    nobs=clicks_nobs,
    alternative="two-sided"
)

print("=== Conversion Rate (CVR) Comparison: Variant A vs B ===")
print(f"CVR A (coupon/discount) = {a_row['cvr']:.4%}")
print(f"CVR B (non-coupon)      = {b_row['cvr']:.4%}")
print(f"z-stat (CVR) = {z_cvr:.3f}, p-value = {p_cvr:.4f}")

# 95% CI for difference in CVR (B - A)
pA_cvr = a_row["cvr"]
pB_cvr = b_row["cvr"]
nA_cvr = a_row["clicks"]
nB_cvr = b_row["clicks"]

diff_cvr = pB_cvr - pA_cvr

se_diff_cvr = np.sqrt(
    pA_cvr * (1 - pA_cvr) / nA_cvr +
    pB_cvr * (1 - pB_cvr) / nB_cvr
)

z_crit = 1.96
ci_low_cvr  = diff_cvr - z_crit * se_diff_cvr
ci_high_cvr = diff_cvr + z_crit * se_diff_cvr

print(f"\nDifference (B - A) in CVR = {diff_cvr:.4%}")
print(f"95% CI for CVR diff = [{ci_low_cvr:.4%}, {ci_high_cvr:.4%}]")
```

### Conversion Rate (CVR)  Two-Proportion Z-Test

Here I test whether **conversion rate** differs between:

- **Variant A** = coupon/discount campaigns  
- **Variant B** = non-coupon campaigns  

Setup:

- Successes = `conversions`  
- Trials   = `clicks`  
- CVR = conversions / clicks  

Hypotheses:

- **H:** p_A = p_B   no difference in CVR  
- **H:** p_A  p_B   CVR differs between A and B  

Results:

- CVR A (coupon/discount)  **8.63%**  
- CVR B (non-coupon)       **13.64%**  
- Difference (B  A)       **+5.02 percentage points**  
- z-stat  **75.43**, p-value  **0.0000**  
- 95% CI for (CVR_B  CVR_A) = **[4.88%, 5.15%]**

Interpretation:

Even though coupon/discount ads attract more clicks (higher CTR),  
**non-coupon campaigns convert a significantly larger share of clicks** into conversions.

So Variant A is a *click magnet*, but Variant B brings **higher-quality traffic** in terms of conversion efficiency, which is crucial when we care about revenue, ROAS, and profit.


## Welch t-test on ROAS (row-level)

```python
from scipy import stats

roas_A = shop_df.loc[shop_df["variant"] == "A", "ROAS"]#.dropna(
roas_B = shop_df.loc[shop_df["variant"] == "B", "ROAS"] #.dropna()

t_stat, p_val = stats.ttest_ind(roas_B, roas_A, equal_var=False)

print(f"Mean ROAS A (coupon/discount) = {roas_A.mean():.3f}")
print(f"Mean ROAS B (non-coupon)      = {roas_B.mean():.3f}")
print(f"Welch t-test: t = {t_stat:.3f}, p-value = {p_val:.4f}")
```

### ROAS Comparison  Welch t-test (Variant A vs B)

Here I compare **ROAS** (Return on Ad Spend) between:

- **Variant A** = coupon/discount campaigns  
- **Variant B** = non-coupon campaigns  

ROAS is defined as:

\[
\text{ROAS} = \frac{\text{Revenue}}{\text{Cost}}
\]

I run a **Welch t-test** on the row-level ROAS values to allow for different variances between the two groups.

**Hypotheses:**

- **H:** _A = _B  mean ROAS is the same for coupon and non-coupon campaigns  
- **H:** _A  _B  mean ROAS differs between the two

**Results:**

- Mean ROAS A (coupon/discount)  **0.86**  
- Mean ROAS B (non-coupon)       **1.13**  
- Welch t-test: **t  1.29**, **p  0.199**

**Interpretation:**

Numerically, non-coupon campaigns have a **higher average ROAS** than coupon/discount campaigns (1.13 vs 0.86).  
However, the p-value (~0.20) is **greater than 0.05**, so I **fail to reject the null hypothesis**.

In other words, given the variability in ROAS across ad groups, I do **not** have strong statistical evidence that coupon campaigns and non-coupon campaigns differ in mean ROAS, even though the point estimates suggest that non-coupon ads are more efficient per dollar spent.


### Note

### Why ROAS differs between the aggregate table and the t-test

Youll notice that ROAS from the A/B aggregate table is not the same as the mean ROAS used in the Welch t-test. Thats expected, because they are two different quantities:

- In the aggregate table, I compute **portfolio-level ROAS** per variant:

  \[
  \text{ROAS}_\text{agg} = \frac{\sum \text{Revenue}}{\sum \text{Cost}}
  \]

  This is a **cost-weighted** metric. Every euro of spend is weighted correctly, so this is the right number to use for business reporting.

- For the Welch t-test, I first compute **row-level ROAS**:

  \[
  \text{ROAS}_i = \frac{\text{Revenue}_i}{\text{Cost}_i}
  \]

  and then take the **simple (unweighted) average** across rows within each variant:

  \[
  \text{Mean ROAS}_A = \frac{1}{n_A} \sum_{i \in A} \text{ROAS}_i
  \]

  Here, each row counts equally, even if some rows have tiny spend and extreme ROAS values. This gives a typical row ROAS, not a total-money-weighted one.

Because of this difference (weighted vs unweighted), the aggregate ROAS
(0.869 for A, 0.903 for B) and the mean row-level ROAS
(0.860 for A, 1.133 for B) do not match, and that is mathematically normal.

In short:
- **Use aggregate ROAS** to talk about overall efficiency per  spent.
- **Use row-level ROAS** as input to statistical tests that compare the distribution of efficiency across ad groups.


## Difference-in-Differences (DiD) on ROAS

```python
import statsmodels.formula.api as smf

# Treated: A (coupon/discount) = 1, B = 0
shop_df["treated"] = np.where(shop_df["variant"] == "A", 1, 0)

# Post: months October & November = 1, July/Aug/Sept = 0
shop_df["Month"] = shop_df["Month"].str.strip().str.capitalize()
post_months = ["October", "November"]
shop_df["post"] = shop_df["Month"].isin(post_months).astype(int)

# Interaction term treated * post
shop_df["treated_post"] = shop_df["treated"] * shop_df["post"]

shop_df.head(3)
```

```python
shop_df[shop_df['ROAS'].isna()]
```

```python
did_model = smf.ols("ROAS ~ treated + post + treated:post", data=did_df).fit()
print(did_model.summary())
```

### Difference-in-Differences on ROAS  Interpretation

Setup:

- Treated group = Variant A (coupon/discount campaigns)
- Control group = Variant B (non-coupon campaigns)
- Pre-period = JulySeptember
- Post-period = OctoberNovember
- Model: `ROAS ~ treated + post + treated:post`

Key coefficients:

- **Intercept (1.10)**  
  Baseline ROAS for non-coupon campaigns (control group) in the pre-period.

- **treated (0.31, p = 0.386)**  
  Before October, coupon campaigns had ROAS about 0.31 lower than non-coupon campaigns, but this gap is not statistically significant.

- **post (+0.08, p = 0.824)**  
  For non-coupon campaigns, ROAS in OctNov is only ~0.08 higher than in JulSep, and this time trend is not significant.

- **treated:post (+0.09, p = 0.868)**   DiD estimate  
  This is the incremental post-period change in ROAS for coupon campaigns relative to non-coupon campaigns. The estimate is small (~0.09) with a very wide 95% CI ( [1.02, 1.21]) and a large p-value, so it is statistically indistinguishable from zero.

**Conclusion:**  
Under this simple DiD specification, there is **no evidence** that coupon/discount campaigns experienced a different ROAS trend over time compared to the other campaigns. Any observed differences across months look like noise rather than a stable causal lift or drop due to being coupon-themed.


##Quick DiD table: predicted ROAS by group & period

```python
# ================================
# Predicted ROAS by (treated, post)
# ================================
import pandas as pd

cells = []
for treated_val in [0, 1]:
    for post_val in [0, 1]:
        mask = (did_df["treated"] == treated_val) & (did_df["post"] == post_val)
        sub = did_df[mask]
        if len(sub) == 0:
            pred_mean = np.nan
        else:
            pred_mean = did_model.predict(sub).mean()
        cells.append({
            "treated (1=A)": treated_val,
            "post (1=Oct/Nov)": post_val,
            "predicted_ROAS": pred_mean
        })

did_cells = pd.DataFrame(cells)
did_cells
```

The table shows:

- Coupon campaigns: ROAS increases from ~0.79 (pre) to ~0.96 (post), but remains below non-coupon campaigns in both periods.

This 22 layout is the concrete version of the DiD model: it shows how ROAS changes over time for treated vs control.  
However, the DiD interaction coefficient (`treated:post`) was small and not statistically significant, so these differences should be interpreted as **weak / suggestive patterns**, not as a strong causal effect of running coupon campaigns.


## Bayesian A/B test  Conversion rate per impression

```python
# Reuse the aggregated ab_agg
a_row = ab_agg[ab_agg["variant"] == "A"].iloc[0]
b_row = ab_agg[ab_agg["variant"] == "B"].iloc[0]

# Use conversions per impression (you can change to per click if you want)
x_A = int(a_row["conversions"])
n_A = int(a_row["impressions"])

x_B = int(b_row["conversions"])
n_B = int(b_row["impressions"])

x_A, n_A, x_B, n_B
```

```python
## Posterior sampling with Beta priors
```

```python
# ================================
# Beta-Binomial posterior sampling
# ================================
# Prior: Beta(1, 1) = uniform (non-informative)

alpha_A, beta_A = 1 + x_A, 1 + n_A - x_A
alpha_B, beta_B = 1 + x_B, 1 + n_B - x_B

samples_A = np.random.beta(alpha_A, beta_A, size=100_000)
samples_B = np.random.beta(alpha_B, beta_B, size=100_000)

lift = samples_B - samples_A

prob_B_better = (lift > 0).mean()
ci_low, ci_high = np.percentile(lift, [2.5, 97.5])

print(f"P(conv_rate_B > conv_rate_A) = {prob_B_better:.3f}")
print(f"95% credible interval for lift (B - A) = [{ci_low:.4%}, {ci_high:.4%}]")
```
